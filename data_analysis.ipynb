```python
import pandas as pd
import altair as alt
```

```python
df=pd.read_csv("data/scraped_laptops_with_specs.csv")
df.head()
```

```python
# Replace '€' and turn price into float object
df['Price'] = df['Price'].str.replace('€', '', regex=False)

df['Price'] = df['Price'].str.replace(',', '.', regex=False)

df['Price'] = pd.to_numeric(df['Price'])
```

```python
df.describe()
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X('brand').title('Brand'),
    alt.Y('count(brand)').title('Number of Records'),
).properties(
    width=600,
    height=400,
    title="Distribution of Brands"
)
```

```python
alt.Chart(df).mark_circle(size=60,opacity=0.7).encode(
    alt.X('brand').title('Brand'),
    alt.Y('Price').title('Price (€)'),
    tooltip=['model_name','Price','CPU','GPU','RAM','disk_space'] #hover over graph
).properties(
    width=600,
    height=400,
    title=("Prices by Brand")
)
```

```python
# Find the index of the row with the minimum price
min_price_index = df['Price'].idxmin()
max_price_index =df['Price'].idxmax()

# Extract the row with the minimum price
cheapest_laptop = df.loc[min_price_index]
priciest_laptop=df.loc[max_price_index]

# Generate the descriptive text
description = (
    f"In the dataset, the cheapest laptop is from {cheapest_laptop['brand']} and costs {round(cheapest_laptop['Price'])}€."
    f" It has a {cheapest_laptop['CPU']} CPU, {cheapest_laptop['RAM']} GB of RAM, and {cheapest_laptop['disk_space']} of disk space. \n"
    f"The most expensive laptop is from {priciest_laptop['brand']} and costs {round(priciest_laptop['Price'])}€."
    f" It has a {priciest_laptop['CPU']} CPU, {priciest_laptop['RAM']} GB of RAM, a {priciest_laptop['GPU']}, and {priciest_laptop['disk_space']} of disk space."
)

print(description)
```
